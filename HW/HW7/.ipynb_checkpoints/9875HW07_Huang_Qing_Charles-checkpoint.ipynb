```r
#(a)
sub_Heston <- list(lbd = 1.3253,rho = -0.7165,eta = 0.3877,vbar = 0.0354,v = 0.0174) 
subSVJ = list(lbd = 0.5394,rho =-0.6974,eta = 0.3038,vbar = 0.0439,v =0.0158,lbdJ = 0.1308,alpha =-0.1151,delta =0.0967) 

# The Heston model
phiHeston <- function(parms){
    
    lbd <- parms$lbd 
    rho <- parms$rho 
    eta <- parms$eta 
    vbar <- parms$vbar 
    v <- parms$v 
    
    function(u, t){

    al <- -u*u/2 - 1i*u/2 
    bet <- lbd - rho*eta*1i*u 
    gam <- eta^2/2 
    d <- sqrt(bet*bet - 4*al*gam) 
    rp <- (bet + d)/(2*gam) 
    rm <- (bet - d)/(2*gam) 
    g <- rm / rp 
    D <- rm * (1 - exp(-d*t))/ (1 - g*exp(-d*t)) 
    C <- lbd * (rm * t - 2/eta^2 * log( (1 - g*exp(-(d*t)))/(1 - g) ) ) 
    return(exp(C*vbar + D*v)) 
    }
}

# phiJump
phiJump <- function(parms){

    lbdJ <- parms$lbdJ  
    alpha <- parms$alpha 
    delta <- parms$delta 

    function(u,t){  
    tmp <- exp(-1i*u*lbdJ*(exp(alpha + delta^2/2) - 1)*t + 
    lbdJ*t*(exp(1i*u*alpha - u^2/2*delta^2) - 1)) 
    return(tmp) 
    }         
}


# The HESJ or SVJ model
phiSVJ <- function(parms){
    function(u, t){  
    return(phiHeston(parms)(u,t)*phiJump(parms)(u,t)) 
    }
}

k <- seq(-1,1,by=0.01)
ImpVolHeston <- sapply(k, function(x){bsvol(phiHeston(sub_Heston),x,1)})
ImpVolSVJ <- sapply(k,function(x){bsvol(phiSVJ(subSVJ),x,1)})

plot(k,ImpVolHeston,type="l",col="blue", xlab="k", ylab="Implied Vol")
lines(k,ImpVolSVJ,type="l",col="red")
```

```r
plotIvols_modified<-function (ivolData, sviMatrix = NULL, slices = NULL) 
{
    bid_volatilitys <- as.numeric(ivolData$Bid)
    ask_volatilitys <- as.numeric(ivolData$Ask)
    expDates <- unique(ivolData$Texp)
    nSlices <- length(expDates)
    if (!is.null(slices)) {
        nSlices <- length(slices)
    }
    else {
        slices <- 1:nSlices
    }
    colnum <- sqrt(nSlices * 2)
    rows <- round(colnum/2, 0)
    columns <- round(colnum, 0)
    while (rows * columns < nSlices) {
        rows <- rows + 1
    }
    atmVol <- numeric(nSlices)
    atmSkew <- numeric(nSlices)
    par(mfrow = c(rows, columns), mex = 0.5)
    for (slice in slices) {
        t <- expDates[slice]
        texp <- ivolData$Texp
        bid_volatility <- bid_volatilitys[texp == t]
        ask_volatility <- ask_volatilitys[texp == t]
        midVol <- (bid_volatility + ask_volatility)/2
        f <- (ivolData$Fwd[texp == t])[1]
        k <- log(ivolData$Strike[texp == t]/f)
        include <- !is.na(bid_volatility)
        k_min <- min(k[include])
        k_max <- max(k[include])
        ybottom <- 0.8 * min(bid_volatility[include])
        ytop <- 1.2 * max(ask_volatility[include], na.rm = T)
        xrange <- c(k_min, k_max)
        yrange <- c(ybottom, ytop)
        plot(k, bid_volatility, col = "red", pch = 18, cex = 0.5, xlim = xrange, 
            ylim = yrange, main = paste("T =", format(t, digits = 2, 
                nsmall = 2)), xlab = "Log-Strike", ylab = "Implied Vol.")
        par(new = T)
        plot(k, ask_volatility, col = "blue", pch = 18, cex = 0.5, xlim = xrange, 
            ylim = yrange, main = NA, xlab = NA, ylab = NA)

        ImpVolHeston <- sapply(k, function(x){bsvol(phiHeston(sub_Heston),x,t)})
        ImpVolSVJ <- sapply(k,function(x){bsvol(phiSVJ(subSVJ),x,t)})
        lines(k,ImpVolHeston,type="l",col="green")
        lines(k,ImpVolSVJ,type="l",col="black")

        if ((!is.null(sviMatrix))) {
            vol <- function(k) {
                sqrt(svi(sviMatrix[slice, ], k)/t)
            }
            par(new = T)
            curve(vol(x), from = k_min, to = k_max, col = "orange", 
                lwd = 2, add = T)
        }
        kIn <- k[!is.na(midVol)]
        volIn <- midVol[!is.na(midVol)]
        volInterp <- function(xout) {
            stinterp(x = kIn, y = volIn, xout)$y
        }
        atmVol[slice] <- volInterp(0)
        atmSkew[slice] <- (volInterp(0.01) - volInterp(-0.01))/0.02
    }
    par(mfrow = c(1, 1), mex = 1)
    par(new = F)
    return(list(expiries = expDates, atmVol = atmVol, atmSkew = atmSkew))
}
res <- plotIvols_modified(spxOptData)
```

## Stochastic time change

### 2. (8 points) 

Consider the following model of the stock price:

$$
S_t = S_0\,\exp\left\{\sigma\,W(\tau_t)-\frac12\,\sigma^2\,\tau_t\right\}
$$

where $W(\tau_t)$ is time-changed Brownian motion.  The time-change $\tau$ is given by

$$
\tau_t = 
\left\{
\begin{array}{ll}
 \theta_1\,t& \text{ with probability } 1/2  \\
\theta_2\,t & \text{ with probability } 1/2.
\end{array}
\right.
$$

Note that the time-change is independent of the Brownian motion $W$.

(a)

Compute the characteristic function 

$$
\phi_{\tau_t}(u) = \ee{e^{i\,u\,\tau_t}}
$$

of the time-change.

(b)

Use equation (5) of Lecture 7 to compute the characteristic function 

$$
\phi_T(u) = \ee{e^{i\,u\,x_T}}= \frac 12 \,\left\{e^{-\frac12\,u\,(u+i)\,\sigma^2\,\theta_1\,T}+ e^{-\frac12\,u\,(u+i)\,\sigma^2\,\theta_2\,T} \right\}.
$$

with $x_t = \log S_t/S_0$.

(c)

Without explicit computation, deduce the fair value $C(S,K,T)$ of a European-style call option with strike $K$ and time to expiry $T$ in this model.

Hint: Note that the characteristic function is the sum of two simple ones.

(d) Rederive your result in part (c) by conditioning on total variance, Hull-White style.

# Answer

(a)
$$
\phi_{\tau_t}(u) = \ee{e^{i\,u\,\tau_t}}= \frac 12 e^{i\,u\,\theta_1\,t}+\frac 12 e^{i\,u\,\theta_2\,t}
$$
(b)

equation (5) is 

\begin{eqnarray*}
\phi_{Y_t}(u)&=&\mathbb{E}\left[e^{i\,u\,Y_t}\right]\nonumber\\
&=&\mathbb{E}\left[\mathbb{E}\left[\left.e^{i\,u\,X(\theta)}\right|\theta=\tau_t\right]\right]\nonumber\\
&=&\mathbb{E}\left[e^{\psi_X(u)\,\tau_t}\right]\nonumber\\
&=&\phi_{\tau_t}\left(-i\,\psi_X(u)\right)
\end{eqnarray*}
$$X=\log S_t/S_0(t) =\sigma\,W_t-\frac12\,\sigma^2\,t $$
$${\mathbb{E}\left[e^{i\,u\,X}\right]}=e^{-\frac12\,u\,(u+i)\,\sigma^2\,}$$
Therefore,

$$
\phi_T(u) = \phi_{\tau_T}\left(i (-i\,\log{\mathbb{E}\left[e^{i\,u\,X}\right]})\tau_T\right)= \frac 12 \,\left\{e^{-\frac12\,u\,(u+i)\,\sigma^2\,\theta_1\,T}+ e^{-\frac12\,u\,(u+i)\,\sigma^2\,\theta_2\,T} \right\}.
$$

(c)

Formula (2.10) of Lewis
$$
C(S,K,T,\phi_T)=S-\sqrt{SK}\frac{1}{\pi}\int_0^\infty\frac{du}{u^2+\frac{1}{4}}
\,\mathrm{Re}\left[e^{-iuk}\phi_T\left(u-i/2\right)\right]
$$
Plug in Black-Scholes' characteristic function, the option value is just Black-Scholes' value.

$C(S,K,T,\phi_T)$ is a linear combination of BS value, which is 

$$C(S,K,T,\phi_T)=\frac 12 C(S,K,\theta_1\,T,\phi_{BS})+\frac 12 C(S,K,\theta_2\,T,\phi_{BS})=\frac 12 C_{BS}(S,K,\theta_1\,T)+\frac 12 C_{BS}(S,K,\theta_2\,T)$$

(d)

\begin{eqnarray*}
C(k,v,T)&=&\mathbb{E}_{w_T}\left[C_{BS}(k,w)\right]\\
&=&\mathbb{E}_{w_T}\left[\mathbb{E}_{S_w}\left[(S_w-K)^+\right]\right]
\end{eqnarray*}

where $w_T=\int_0^T\,v_t\,dt$
 

$$
w_t = 
\left\{
\begin{array}{ll}
\sigma^2\theta_1\,T& \text{ with probability } 1/2  \\
\sigma^2\theta_2\,T & \text{ with probability } 1/2.
\end{array}
\right.
$$

Therefore, 

$$C(S,K,T)=\frac 12 C_{BS}(S,K,\theta_1\,T)+\frac 12 C_{BS}(S,K,\theta_2\,T)$$
