$$
\begin{eqnarray*}
\frac{dS_t}{S_t} &=& \sqrt{v_t}\,S_t \,dZ_t\\
dv_t &=&  - \kappa \,(v_t - \bar v)\,dt + \alpha \,v_t \,dW_t
\end{eqnarray*}
$$

with $
\left\langle {dZ_t \,{dW_t} } \right\rangle = \rho \, dt

mean $\bar v$.

(b)
